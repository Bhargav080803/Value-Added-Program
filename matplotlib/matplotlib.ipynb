```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=np.array([3,8,12,14,15,24,38,42,49,55])
y=np.array([11,16,34,47,61,67,69,75,88,91])
z = np.array([76,74,66,54,43,32,26,28,21,13])
p= np.array([56,66,44,34,67,66,52,74,37,66])
```

#### Line plot

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(x,p)
plt.show()
```

```python
plt.plot(y,z)
plt.show()
```

```python
plt.title('My first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x,y,color ='red',marker='o',mfc='green',lw=5)
#Marker :.o > < v ^ * + D
```

```python
df=pd.read_csv('student.csv')
```

```python
plt.title('roll vs maks')
plt.xlabel('roll number')
plt.ylabel('Marks')
plt.grid()
plt.xlim(0,11)

plt.ylim(40,100)
plt.yticks(range(40,100,5))

plt.plot(df['roll'],df['marks'],color ='blue',marker='o',mfc='green')
#Marker :.o > < v ^ * + D
```

```python
plt.title('My first plot')
plt.xlabel('first value')
plt.ylabel('second value')
plt.grid()
plt.xlim(0,70)
plt.ylim(0,100)
plt.yticks(range(0,100,10))
plt.plot(x,y,color ='red',label='2023')
plt.plot(x,z,color ='green',label='2024')
plt.legend()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,color ='green')
plt.subplot(1,2,2)
plt.plot(x,z,color='r')
```


#### scatter plot
