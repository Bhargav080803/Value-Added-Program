```python
import urllib
```

```python
url= urllib.request.urlopen('https://en.wikipedia.org/wiki/Rajgad_Fort')
html=url.read()
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup= BeautifulSoup(html,'html')
```

```python
text= soup.get_text(strip=False)
```

```python
print(text)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens1= word_tokenize(text)
```

```python
from nltk import FreqDist
freq=FreqDist(tokens1)
```

```python
freq.plot(20)
```

```python
from nltk.corpus import stopwords
import string
```

```python
swords=stopwords.words('english')
#punctuation removal
tokens2=[x for x in tokens1 if x.isalpha()]
```

```python
tokens3=[x.lower() for x in tokens2 if x.lower() not in swords] 
```

```python
freq= FreqDist(tokens3)
```

```python
freq.plot(20,marker='d',)
```

```python
freq.most_common(21)
```

Stemming


```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
tokens4 =[ps.stem(x)for x in tokens3]
```

```python
freq = FreqDist(tokens4)
freq.plot(20,marker='d',color='b')
```

```python
url= urllib.request.urlopen('https://en.wikipedia.org/wiki/Raigad_Fort')
html=url.read()

soup= BeautifulSoup(html,'html')
text= soup.get_text(strip=False)
tokens1= word_tokenize(text)

#punctuation removal
tokens2=[x for x in tokens1 if x.isalpha()]
tokens3=[x.lower() for x in tokens2 if x.lower() not in swords] 

tokens4 =[ps.stem(x)for x in tokens3]
freq = FreqDist(tokens4)
freq.most_common(3)
```
